2) What is Daniel Jones’ definition of a phoneme?
- A family of uttered sounds (segmental elements of speech) in a particular language which count for practical purposes as if they were one and the same.


3) What is the phoneme restoration effect?
- If a short section of speech was cut out and replaced by another sound (such as a cough), listeners could not detect that anything was missing; the excised sound was restored in the mind of the listener.

9) In section 3.3 the authors give a list of examples of the misuse of the term “phoneme” The author gives a short explanation for each that is not comprehensive. Choose 3 examples and explain fully why each is a misuse. 
- "We have 252 phonemes, of which there are 213 Mandarin and 39 English." One language only has a limited set of phonemes, but here it seems that it's trying to use it to mean the count of each phone in a given dataset.
- "Spectral transitions between phonemes... " Phonemes are abstract psychological concepts, not the realisation of a sound, so there are no transitions between phonemes, only phones.
- "We propose a language-independent phoneme segmentation method.” Phonemes are sets of groupings of sounds in the head of a native speaker, and this set differs language by language.
